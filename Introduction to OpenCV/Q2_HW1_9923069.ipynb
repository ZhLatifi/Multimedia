```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow function 
def imshow(title= "", image= None, size= 8.5, axixEn= False):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    if axixEn == False:
      plt.tick_params(left= True, labelleft= True , labelbottom= True, bottom= True)
    plt.show()

# Load our input image
image = cv2.imread('C:/Users/USER/Desktop/Zh/University/Term6/Multimedia/HW/HW1/img2.png')

# Show the image with matplotlib
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Use cv2.split to get each color space separately
B, G, R = cv2.split(image)
imshow("Blue Channel Only", B)
imshow("Green Channel Only", G)
imshow("Red Channel Only", R)
```

```python
# If no interpolation is specified cv.INTER_LINEAR is used as default
# Let's make our image 1/2 of it's original size
image_scaledB = cv2.resize(B, None, fx=0.2, fy=0.2)
imshow("0.2x Scaling Blue - Linear Interpolation", image_scaledB)

# Show the plotting graph of an image
histrB = cv2.calcHist([image_scaledB],[0],None,[256],[0,256])
plt.plot(histrB, 'b')
plt.title("Histogram of blue channel after scaling")
plt.show()
```

```python
# If no interpolation is specified cv.INTER_LINEAR is used as default
# Let's make our image 3x of it's original size
image_scaledG = cv2.resize(G, None, fx=0.2, fy=0.2)
imshow("0.2x Scaling Grean - Linear Interpolation", image_scaledG)

# Show the plotting graph of an image
histrG = cv2.calcHist([image_scaledG],[0],None,[256],[0,256])
plt.plot(histrG, 'g')
plt.title("Histogram of green channel after scaling")
plt.show()
```

```python
# If no interpolation is specified cv.INTER_LINEAR is used as default
# Let's make our image 1/10 of it's original size
image_scaledR = cv2.resize(R, None, fx=0.2, fy=0.2)
imshow("0.2x Scaling Red - Linear Interpolation", image_scaledR)

# Show the plotting graph of an image
histrR = cv2.calcHist([image_scaledR],[0],None,[256],[0,256]) 
plt.plot(histrR, 'r')
plt.title("Histogram of red channel after scaling")
plt.show()
```

```python
# Show the plotting histogram of all images

plt.plot(histrR, 'r')
plt.plot(histrG, 'g')
plt.plot(histrB, 'b')
plt.title("Histograms of three channels after scaling")
plt.show()
```

```python
# Merge new channels and re-make the original image
merged = cv2.merge([image_scaledB, image_scaledG, image_scaledR]) 
imshow("Merged image after scaling", merged)

# Show the plotting graph of an image
histrMrg = cv2.calcHist([merged],[0],None,[256],[0,256])
histrOri = cv2.calcHist([image],[0],None,[256],[0,256])
plt.plot(histrMrg)
plt.plot(histrOri, 'r')
plt.title("Histograms of Original image and Merged image after scaling")
plt.show()
```
