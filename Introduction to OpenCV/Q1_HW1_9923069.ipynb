```python
image.astype(float)
image = image / 255
image2 = image.copy()

for i in range(0, image.shape[0]):
    for j in range(0, image.shape[1]):
        if (image[i][j][0] < 0.55) and (image[i][j][0] > 0.2):
            image2[i][j][:] = 0
            
plt.imshow(image2)
plt.show()
```

```python
image3 = image.copy()

for i in range(0, image3.shape[0]):
    for j in range(0, image3.shape[1]):
        if (image[i][j][0] < 0.4):
            image3[i][j][:] = 0
        elif (image[i][j][0] > 0.55):
            image3[i][j][:] = 1

            
plt.imshow(image3)
plt.show()
```

```python
image4 = image.copy()

for i in range(0, image4.shape[0]):
    for j in range(0, image4.shape[1]):
        if (image[i][j][0] < 0.2):
            image4[i][j][:] = float(3/2)*image[i][j][:]           
        elif (image[i][j][0] > 0.55):
            image4[i][j][:] = float(14/9)*image[i][j][:] - float(5/9)
        else:
            image4[i][j][:] = 0.3 

plt.imshow(image4)
plt.show()
```
