```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow function 
def imshow(title= "", image= None, size= 8.5, axixEn= False):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    if axixEn == False:
      plt.tick_params(left= True, labelleft= True , labelbottom= True, bottom= True)
    plt.show()

# Load our input image
image = cv2.imread('C:/Users/USER/Desktop/Zh/University/Term6/Multimedia/HW/HW1/img4.jpg')

# Show the image with matplotlib
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Using Median Filter

median = cv2.medianBlur(image, 11)
imshow("Median Blurring", median)
```

```python
# Using Gaussian Filter

Gaussian = cv2.GaussianBlur(image, (13, 13), 0)
imshow("Gaussian Blurring", Gaussian)
```

```python
# Using Bilateral Filter

bilateral = cv2.bilateralFilter(image, 11, 500, 500)
imshow("Bilateral Blurring", bilateral)
```

```python
# Define Our Soble Function & Display Sobel Edge Detection

def Sobel(input):
    gray = cv2.cvtColor(input, cv2.COLOR_BGR2GRAY)
#     gray = input
    sobelx = cv2.Sobel(src=gray, ddepth=cv2.CV_8U, dx=1, dy=0, ksize=3) # Sobel Edge Detection on the X axis
    sobely = cv2.Sobel(src=gray, ddepth=cv2.CV_8U, dx=0, dy=1, ksize=3) # Sobel Edge Detection on the Y axis
    sobelxy = cv2.Sobel(src=gray, ddepth=cv2.CV_8U, dx=1, dy=1, ksize=3) # Combined X and Y Sobel Edge Detection
    return [sobelx, sobely, sobelxy]

 
# Original Image
imshow("Original image's sobel X", Sobel(image)[0]) 
imshow("Original image's sobel Y", Sobel(image)[1]) 
imshow("Original image's sobel XY", Sobel(image)[2])

# Median
imshow("Median's sobel X", Sobel(median)[0]) 
imshow("Median's sobel Y", Sobel(median)[1]) 
imshow("Median's sobel XY", Sobel(median)[2])
 
#Gaussian
imshow("Gaussian's sobel X", Sobel(Gaussian)[0]) 
imshow("Gaussian's sobel Y", Sobel(Gaussian)[1]) 
imshow("Gaussian's sobel XY", Sobel(Gaussian)[2])
 
# Bilateral
imshow("Bilateral's sobel X", Sobel(bilateral)[0]) 
imshow("Bilateral's sobel Y", Sobel(bilateral)[1]) 
imshow("Bilateral's sobel XY", Sobel(bilateral)[2])
```

```python
# Define Our Canny Function & Display Canny Edge Detection

def Canny(input):
    edges = cv2.Canny(image=input, threshold1=100, threshold2=200)
    return edges


# Original Image
imshow("Original image's Canny Edge Detection", Canny(image))

# Median
imshow("Median's Canny Edge Detection", Canny(median))

# Gaussian
imshow("Gaussian's Canny Edge Detection", Canny(Gaussian))

# Bilateral
imshow("Bilateral's Canny Edge Detection", Canny(bilateral))
```
